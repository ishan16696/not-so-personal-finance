## Net personal cashflow

Let's first understand some terms.
- **Total income**: As the name suggest, it refers to yours total monthly income from all sources after taxes.
- **Fundamental costs**: Fundamental costs refers to all cost essential for your living. It includes Mobile, Internet and Electricity bills, all EMIs (car and home loan EMIs), Rent, Groceries, Travel/Transportation costs etc.
- **Net cashflow**: It's a measure of how much money is coming in and going out. Your net cash flow is remaining amount after deducting all fundamental expenses.

#### Net cashflow = Total income - Fundamental costs


##### Before moving further here's some rules which I personally follow:
let's first denote terms to make math equations look simple.
1. `Net Personal Cashflow` as ${C}$, `Total income` as ${T}$ and `Fundamental costs/expenses` as ${F}$.
2. Now, all `EMIs` as ${e}$ and rest of remaining of the `total expenses` as ${x}$.
3. So, `Fundamental costs` is sum of EMIs and rest of the total expenses. ${F=e+x}$.


So, the equation of total cashflow becomes: ${C=T-F}$ or ${C=T-(e+x)}$ <br/>
There are some rules which I personally flows:

### Rules to keep in mind

1. ${C >= T/2}$ it says your net cash flow should be greater than ${50}$% of your total income.
2. ${e <= T/4}$ it says your all EMIs must be less than ${25}$% of your total income.
3. ${x <= T/4}$ it says your rest of expenses(excluding EMIs) must be less than ${25}$% of your total income.
4. ${F <= T/2}$ it says your fundamental costs of living must be less than ${50}$% of your total income.
5. ${\dfrac{dx}{dt} < \dfrac{dT}{dt}}$ , it says rate at which your expenses will grow should be less than rate at which your total income will grow.

$\color{blue}{\text{You can follow these rules to determine whether you can afford a particular expense, and adjust them as necessary to fit your financial situation.}}$


**Note**: Your $\color{red}{\text{net personal cashflow}}$ is important as this amount will be used for your future financial planning. We will talk about this in details later.

> Note: These rules are particularly relevant for younger individuals, as they often have significant portions of their expenditures dedicated to EMIs(house/car loan).

### Emergency Fund

In my opinion, in today's unpredictable world, the necessity of an `Emergency Fund` can't be overlooked. While the general recommendation is to maintain an emergency fund equivalent to six months of expenses, I would recommend going a step further and maintain an emergency fund that covers 12 months of expenses.

**Emergency Fund** = ${12}$ months worth of EMIs + ${12}$ months worth of remaining expenses


### Financial Freedom

- I define financial freedom as having sufficient cash flow to cover your basic fundamental expenses for several years, let say for 4 to 5 years. This will provide a safety net against job layoffs and even allows for a sabbatical to explore business opportunities.
- It can also be calculated by multiply your annual expenses by ${x}$ no. of years where x should be less than 5 years.

   ${Total Corpus}$ = ${x * {Annual Expenses}}$
   

> Note: This should be separate from 12 months of emergency fund.

### Retirement Planning

- For retirement or semi retirement, I'm considering 25 to 30 years of lifespan left
- You should have a sufficient corpus in your investments and savings to live atleast ~30 years of your life.
- To caculate the total corpus required there are various methods available:
  
  i) Total corpus is inversely proportional to the rate of withdrawl which will cover your annual expenses.

  ${Total Corpus}$ = ${\dfrac{Annual Expenses}{3\%}}$
  
  ii) It can also be calculated by multiply your annual expenses by ${25}$ or ${30}$ to determine the amount needed for passive retirement or financial freedom.

   ${Total Corpus}$ = ${25 * {Annual Expenses}}$

- In market there is rule known as [4% rule](https://www.investopedia.com/terms/f/four-percent-rule.asp) of the Systematic Withdrawal Plan (SWP) which says we can withdraw ${4\%}$ of total corpus annually, but I recomment you to use ${3\%}$ rule of SWP to be on safer side, so you can withdraw ${3\%}$ of total corpus annually to cover your annual expenses.

> Note: Remember to halt the SWP if the market crashes or become very volatile. To handle such situations, diversify your investments into less volatile assets, so you can withdraw money from those assets classes.

## Asset Allocation

### Rules to keep in mind

> Note: You need to invest in $\color{blue}{\text{assets}}$ to outperform atleast inflation.

1. If Goal time is less than 3 years.

   - Then your target should not be grow your wealth, your target should only be beat the inflation.
   - Debt/FDs: ${50}$%
   - Cash or equivalents: ${20}$%
   - Gold or equivalents: ${20}$%
   - Equity: ${10}$% (Index funds or large cap)

2. If Goal time is b/w 4 to 6 years.

   - In this your taget should be balance b/w preserve and grow your wealth.
   - Debt/FDs: ${20}$%
   - Cash or equivalents: ${10}$%
   - Gold or equivalents: ${10}$%
   - Equity: ${60}$% (Index funds)

   
4. If Goal time is greater than 6 years.

   - Then your target should only be grow your wealth.
   - Debt/FDs: ${0}$%
   - Cash or equivalents: ${10}$% (to take advantage of bear market)
   - Gold or equivalents: ${0}$% (no need to hedge as you have already invested in Gold in above 2 stages).
   - Equity: ${90}$%


Let's discuss the asset allocation for all the stages which we discussed above.

#### Emergency Fund

- Divide the total amount of `Emergency Fund` into two parts. Allocate the first half, which equates to six months of EMIs and expenses, to a low-volatility liquid asset, such as a savings account. But remember, it should in a separate saving account.
- The remaining half can be invested in Gold ETFs or FDs or Debt etc., which are highly liquid assets, beat the inflation and exhibit relatively low volatility.


#### Financial freedom
